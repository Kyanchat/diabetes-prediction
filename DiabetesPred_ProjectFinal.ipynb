```python
data_clean['Diabetes_target'] = data_clean['Diabetes_target'].astype('float64')
```

```python
def outliers():
    l = [ 'HighBP', 'HighChol', 'CholCheck', 'BMI', 'Smoker',
       'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
       'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age', 'Education',
       'Income']
    for i in l:
        x = np.quantile(data_clean[i],[0.25,0.75])
        iqr = x[1]-x[0]    # Interquartile range
        lof = x[0]-1.5*iqr   # lower_fence
        upf = x[1]+1.5*iqr   # upper_fence
        data_clean[i] = np.where(data_clean[i]>upf,upf,(np.where(data_clean[i]<lof,lof,data_clean[i])))
outliers()
```

```python
# checking again to see if outliers persists after removal
out = 1
plt.figure(figsize=(20,20))
for i in ['HighBP', 'HighChol', 'CholCheck', 'BMI', 'Smoker',
       'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
       'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age', 'Education','Income']:
    if out<23:
        plt.subplot(6,5,out)
        plt.boxplot(data_clean[i])
        plt.title(i)
        out=out+1
```

## Over Sampling

```python
# over sampling of the dataset to get a balanced dataset
value_0 = data_clean[data_clean['Diabetes_target'] == 0]
value_1 = data_clean[data_clean['Diabetes_target'] == 1]

# over sampling of the minority value 1
value_1_over = value_1.sample(len(value_0), replace=True)

# Creating a new dataframe with over sampled value 1 data and value 0 data
data_sampled = pd.concat([value_1_over, value_0], axis=0)

# plotting the new label distribution
data_sampled['Diabetes_target'].value_counts().plot(kind='bar', title='value Distribution after Oversampling')
```

```python
#assigning X to the new sampled data
X = data_sampled.drop('Diabetes_target', axis = 1)
X
```

```python
#assigning y to the new sampled data
y = data_sampled['Diabetes_target']
y
```

## Building an Artificial Neural Network (ANN) model using Keras

```python
 pip install tensorflow
```

```python
import tensorflow as tf
#creating sequential model
model = tf.keras.models.Sequential()
#adding a first dense layer with 512 neurons to the model
#activation function relu for non linearity
#shape 21 number of features
model.add(tf.keras.layers.Dense(units=512, activation='relu', input_shape = (21, )))
model.add(tf.keras.layers.Dropout(0.2))

#adding another dense layer with 512 neurons to the model
#activation function relu 
model.add(tf.keras.layers.Dense(units=512, activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))

#adding another dense layer with 100 neurons to the model
#activation function relu
model.add(tf.keras.layers.Dense(units=100, activation='relu'))
#Add dropout layer after each dense layer except the output layer
model.add(tf.keras.layers.Dropout(0.2))
          
#Adding final output layer with 1 neuron
#Activation function sigmoid produces a probability value between 0 and 1, suitable for binary classification tasks
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid')) 
```

```python
#printing summary of the neural network model
model.summary()
```

## Compiling and training the ANN model

```python
#compiling defined neural model with optimizer Adam
#loss function binary_crossentropy
#metrics for evaluation accuracy
model.compile(optimizer='Adam', loss='binary_crossentropy',metrics = ['accuracy'])
```

```python
hist = model.fit(X_train, y_train, batch_size=32, epochs= 100)
```

```python
#feeding test data into X_test and getting model predictions for each data point
y_pred = model.predict(X_test)
```

```python
#converting predictions to classification labels
y_pred = (y_pred > 0.5)
y_pred 
```

## Training a K-Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#create empty lists
test_scores = []
train_scores = []

#iterate loop to test different k values
for i in range(1,20):

#create knn classifier
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
#evaluate the performance    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
print(train_scores, test_scores)
```

```python
## score that comes from testing on the same datapoints that were used for training
maxtrain_score = max(train_scores)
trainscores_ind = [i for i, v in enumerate(train_scores) if v == maxtrain_score]
print('Max train score {} % and k = {}'.format(maxtrain_score*100,list(map(lambda x: x+1, trainscores_ind))))
## score that comes from testing on the datapoints that were split in the beginning to be used for testing solely
maxtest_score = max(test_scores)
testscores_ind = [i for i, v in enumerate(test_scores) if v == maxtest_score]
print('Max test score {} % and k = {}'.format(maxtest_score*100,list(map(lambda x: x+1, testscores_ind))))
```

```python
#visualizing how the training and testing scores change over eppochs
plt.figure(figsize=(12,5))
sns.lineplot(train_scores,marker='*',label='Train Score')
sns.lineplot(test_scores,marker='o',label='Test Score')
```
